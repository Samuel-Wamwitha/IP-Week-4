# Analysis

Research Question

- Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```
#The following lines of code finds the most popular hour of the day for picking up a shared electric blue car in Paris. 

popular_hour = autolib['Bluecar counter'].groupby([autolib['hour'], autolib['City'] == 'Paris']).count()
popular_hour
```
